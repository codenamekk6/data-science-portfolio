ðŸ§  **1. Descriptive Statistics**


```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Create a sample dataset
data = {
    "Test Scores": [67, 85, 90, 76, 88, 92, 70, 78, 82, 94]
}
df = pd.DataFrame(data)

# Summary statistics
print("Descriptive Statistics:\n", df.describe())

# Manual calculations
mean_score = np.mean(df["Test Scores"])
median_score = np.median(df["Test Scores"])
std_score = np.std(df["Test Scores"], ddof=1)

print(f"\nMean: {mean_score}")
print(f"Median: {median_score}")
print(f"Standard Deviation: {std_score}")
```

ðŸ“Š **2. Data Visualization**

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))

# Histogram
plt.subplot(1, 2, 1)
sns.histplot(df["Test Scores"], kde=True, bins=5)
plt.title("Histogram of Test Scores")

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(x=df["Test Scores"])
plt.title("Boxplot of Test Scores")

plt.tight_layout()
plt.show()
```

### ðŸ“Š Interpretation of Histogram

- The histogram shows the distribution of student test scores.
- Most of the scores fall between **85 and 95**, making it the most populated range.
- The curve (kernel density estimate) suggests a **slightly right-skewed** distribution, indicating a small tail on the lower score side.
- There are visible clusters around **80 to 95**, suggesting that many students scored above average.

---

### ðŸ“¦ Interpretation of Boxplot

- The boxplot provides a summary of the test scores using the five-number summary.
- The **median score** is around **85**, which means 50% of the students scored below and 50% above this value.
- The **interquartile range (IQR)**, representing the middle 50% of the data, spans from around **75 to 90**.
- The whiskers extend approximately from **68 to 94**, showing the full range of scores.
- There are **no outliers** in the dataset, as there are no points beyond the whiskers.

---

### ðŸ§  Overall Insight

The test scores are mostly **concentrated in the higher ranges**, suggesting good performance among students. The absence of outliers and the relatively symmetrical shape of the distribution indicate a consistent scoring pattern across the group.


ðŸŽ² **3. Probability Distributions**

```python
# Import necessary library
from scipy.stats import norm

# Normal distribution curve
x = np.linspace(60, 100, 100)
y = norm.pdf(x, loc=mean_score, scale=std_score)

plt.plot(x, y)
plt.title("Normal Distribution of Test Scores")
plt.xlabel("Score")
plt.ylabel("Probability Density")
plt.grid(True)
plt.show()
```

### ðŸ”” Interpretation of Normal Distribution Plot

- This plot represents a **normal (bell-shaped) distribution** using the mean and standard deviation of the test scores.
- The curve peaks around **82â€“84**, which is the **mean** of the dataset.
- The **symmetrical shape** of the curve indicates that the data follows a roughly **normal distribution**.
- Most scores are clustered around the mean, and fewer students scored very low or very high.
- The probability density decreases smoothly as scores move further from the mean, which is typical in naturally occurring data.

---

### ðŸ“Œ Statistical Insight

A normal distribution suggests that:
- The dataset is **well-balanced** without heavy skew.
- **Parametric statistical tests** (e.g., t-test, ANOVA) are appropriate because many of them assume normally distributed data.
